```python
import pandas as pd
```

```python
# open weather maps data
data_1 = pd.read_csv("open_weather_cleaned.csv")
```

```python
# met eireann data
data_2 = pd.read_csv("met_eireann_historical_cleaned.csv")
```

```python
data_1.head()
```

```python
data_2.head()
```

```python
data_1.rename(columns={"dt":"date"}, inplace=True)
data_1["date"] = pd.to_datetime(data_1["date"], unit="s")
```

```python
data_1.head()
```

```python
data_2.rename(columns={"temp":"temp_met_eir"}, inplace=True)
```

```python
data_2["date"] = pd.to_datetime(data_2["date"])
data_2.head()
```

```python
combined = pd.merge(data_1, data_2, on="date")
```

```python
combined.head()
```

```python
k = []
for i in combined.index:
    t1 = combined.loc[i, "temp"]
    t2 = combined.loc[i, "temp_met_eir"]
    
    k.append(abs(t1 - t2))

print(max(k))
print(min(k))
```

```python
j = 0
for i in combined.index:
    main = combined.loc[i, "weather_main"]
    rain = combined.loc[i, "rain"]
    
    if (main == "Rain") and (rain <= 0):
        j += 1

print(j) 
```

```python
h = 0
for i in combined.index:
    main = combined.loc[i, "weather_main"]
    rain = combined.loc[i, "rain"]
    
    if (main != "Rain") and (rain > 0):
        h += 1

print(h)
```

```python
def someFunc(num_array):
    
    n = 1
    m = max(num_array)
    
    while n < m:
        tot = 0
        for i in num_array:
            if i > n:
                tot +=1
        print("%d: %d" % (n, tot))
        n += 1
```
